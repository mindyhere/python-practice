```python
# C(cost) : 정규화 매개변수, 오차에 대한 패널티. C가 작으면 패널티가 커지고 C가 커지면 패널티가 줄어드는 방식
# C값을 증가 → 모델이 좀더 복잡해짐(좀더 자세하게 결정경계를 찾아내는 방식)
# C값이 작으면? 이상치들이 있을 가능성을 크게 잡아서 일반적인 결정 경계를 만들어낸다

# gamma값을 증가 → 각각의 데이터 포인트가 영향력을 행사하는 거리가 감소
# gamma값이 작으면? 영향력을 행사하는 거리가 증가

# cost와 gamma가 너무 크면 과대적합의 위험이 있고 너무 작으면 과소적합의 위험이 있음
```

```python
import matplotlib.pyplot as plt
import mglearn

fig, axes = plt.subplots(3, 3, figsize=(15, 10))

# cost : 0.1, 10, 1000
# gamma : 0.1, 1, 10

for ax, C in zip(axes, [-1, 1, 3]):
    for a, gamma in zip(ax, [-1, 0, 1]):
        mglearn.plots.plot_svm(log_C=C, log_gamma=gamma, ax=a)

# ncol=4 4개의 셀을 합쳐서 범례 표시
axes[0, 0].legend(
    ["class 0", "class 1", "class 0 support vectors", "class 1 support vectors"],
    ncol=4,
    loc=(1, 1.2),
)
```
