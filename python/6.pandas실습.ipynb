```python
# numpy : 벡터 및 행렬 연산에 특화된 라이브러리
import numpy as np

items = range(10, 51, 10)

# numpy 배열 생성
arr = np.array(items)
print(arr)
print(arr.shape)  # 차원
```

```python
# 2차원 numpy 배열
items2 = [range(1, 6), range(10, 51, 10)]
arr2 = np.array(items2)
print(arr2)

# 2행 5열의 2차원 배열
print(arr2.shape)
```

```python
# numpy 배열의 연산
arr1 = np.array([[70, 80], [195, 185]])
arr2 = np.array([[50, 60], [70, 80]])
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
print(arr1 // arr2)
```

```python
student1 = [60, 70, 99]
student2 = [77, 88, 80]
points = np.array([student1, student2])
print(points)
```

```python
print(points.sum())  # 전체합계
print(points.sum(axis=0))  # 열의 합계
print(points.sum(axis=1))  # 행의 합계
print(points.mean())  # 전체평균
print(points.mean(axis=0))  # 열의 평균
print(points.mean(axis=1))  # 행의 평균
```

```python
# 각 요소 중 큰값을 뽑아 새로운 배열 생성
arr = np.maximum(student1, student2)
print(arr)
print(arr.sum())  # 합계
print(arr.mean())  # 평균
```

```python
print(np.sort(arr))  # 오름차순
print(np.sort(arr)[::-1])  # 내림차순
```

```python
# Pandas

# Series : 1차원데이터
# DataFrame : 2차원 표 형태의데이터

import pandas as pd

# Series자료형 생성
sr = pd.Series(range(5, 51, 10))
print(sr)

# 기본 숫자 인덱스 대신 문자 인덱스 사용가능
sr = pd.Series(range(5, 51, 10), index=["a", "b", "c", "d", "e"])
print(sr)
```

```python
emp = {"김철수": 5000, "김철호": 7000, "한상민": 4000, "문대용": 4500}
sr = pd.Series(emp)
print(sr)
# key : value 구조 확인
```

```python
a = pd.DataFrame(
    {
        "name": ["김철수", "김대희", "송영민", "이희연", "정미숙"],
        "weight": [80.0, 70.4, 65.5, 45.9, 51.2],
        "height": [170, 180, 155, 143, 154],
        "gender": ["m", "m", "m", "f", "f"],
    }
)

print("체중")
print(a["weight"])
print("신장과 체중")
print(a[["weight", "height"]])
print("신장 170 이상")
print(a[a.height >= 170])
print("성별 m")
print(a[a.gender == "m"])
print("키 오름차순")
print(a.sort_values(by="height"))
print("체중 내림차순")
print(a.sort_values(by="weight", ascending=False))
```
